# Matplotlib的基本操作

```python
# 导入matplotlib的pyplot包
from matplotlib import pyplot as plt
```

```python
# x轴和y轴数据
x = range(2, 26, 2)
y = [15, 13, 14.5, 17, 20, 25, 26, 26, 24, 22, 18, 15]
```

```python
# 绘制图
plt.plot(x, y)
# 将绘制的图显示出来
plt.show()
```

```python
# 添加标签
plt.xlabel("时间")  # x轴标签
plt.ylabel("温度(℃)")  # y轴标签
plt.plot(x, y)
```

```python
#解决中文不显示的问题
import matplotlib

font = {
    'family': 'WenQuanYi Micro Hei',
    'weight': 'bold',
    'size': 10
}
matplotlib.rc('font', **font)
```
